```python
import numpy as np
import pandas as pd
from statsbombpy import sb
```

```python
data = sb.competitions()
data_matches = sb.matches(competition_id=43,season_id=3)
data_france = data_matches.loc[(data_matches['home_team'] == 'France') | (data_matches['away_team'] == 'France')]

id_matchs = data_france['match_id'].unique()
print(id_matchs)

for index, id in enumerate(id_matchs):

  home_team = data_france[data_france['match_id'] == id]['home_team'].to_string(index=False)
  away_team = data_france[data_france['match_id'] == id]['away_team'].to_string(index=False)

  print("\n\nMatch n°", index, "  id_match: ", id)
  print("home_team: ", home_team, " | away_team: ", away_team)
  data_match = sb.events(match_id=id)

  specific_events = data_match[['team','type','player','player_id','location','pass_recipient', 'pass_outcome', 'pass_type']]
  specific_events = specific_events[(specific_events['team']=='France') & (specific_events['type']=='Pass')]
  specific_events = specific_events[(specific_events['pass_outcome'] != 'Incomplete') & (specific_events['pass_outcome'] != 'Out') & (specific_events['pass_outcome'] != 'Pass Offside') & (specific_events['pass_outcome'] != 'Injury Clearance') & (specific_events['pass_outcome'] != 'Unknown')]
  specific_events = specific_events[(specific_events['pass_type'] != 'Throw-in') & (specific_events['pass_type'] != 'Free Kick') & (specific_events['pass_type'] != 'Kick Off')]

  list_players = specific_events['player'].unique()
  hum = specific_events.groupby('player')

  for player in list_players:
    player_grouped = hum.get_group(player)
    mean_loc = np.zeros(shape=(2,))
    for row in player_grouped['location']:
      mean_loc[0] += row[0]
      mean_loc[1] += row[1]
    mean_loc /= player_grouped.shape[0]
    mean_loc[0] = int(mean_loc[0])
    mean_loc[1] = int(mean_loc[1])

    print("     ", player, ":    mean_loc = ", mean_loc)
```
